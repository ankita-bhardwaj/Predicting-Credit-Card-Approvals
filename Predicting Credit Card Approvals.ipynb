```python
import pandas as pd
cc_apps=pd.read_csv('cc_approvals.data',sep=',',header=None)
```

```python
df=cc_apps.copy()
```

```python
cc_apps_description = cc_apps.describe()
print(cc_apps_description)

print("\n")

cc_apps_info = cc_apps.info()
print(cc_apps_info)

print("\n")

cc_apps.tail(10)
```

```python
cc_apps=df.copy()
import numpy as np

print(cc_apps.tail(17))

cc_apps = cc_apps.replace('?',np.NaN)

print(cc_apps.tail(17))
cc_apps.isna().sum()
```

```python
cc_apps=cc_apps.dropna()
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
for col in cc_apps:

    if cc_apps[col].dtypes=='object':

        cc_apps[col]=le.fit_transform(cc_apps[col])
        
```

```python
from sklearn.model_selection import train_test_split
cc_apps = cc_apps.drop([11, 13], axis=1)
cc_apps = cc_apps.values
X,y = cc_apps[:,0:12] , cc_apps[:,13]

X_train, X_test, y_train, y_test = train_test_split(X,
                                y,
                                test_size= 0.33,
                                random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
rescaledX_train = scaler.fit_transform(X_train)
rescaledX_test = scaler.fit_transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression ()
logreg.fit(rescaledX_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix
y_pred = logreg.predict(rescaledX_test)
```

```python
logreg.score(rescaledX_test,y_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.model_selection import GridSearchCV 
tol=[0.01, 0.001,0.0001]
max_iter=[100, 150,200]
```

```python
param_grid = dict({'tol':tol, 'max_iter':max_iter})
```

```python
grid_model = GridSearchCV(estimator=logreg, param_grid=param_grid, cv=5)
rescaledX = scaler.fit_transform(X)

grid_model_result = grid_model.fit(rescaledX,y)
```

```python
best_score, best_params = grid_model_result.best_score_, grid_model_result.best_params_
print("Best: %f using %s" % (best_score, best_params))
```

```python
for col in cc_apps:

    if cc_apps[col].dtypes == 'object':

        cc_apps = cc_apps.fillna(cc_apps[col].value_counts().index[0])

cc_apps.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
```

```python
for col in cc_apps:
    if cc_apps[col].dtypes=='object':
        cc_apps[col]=le.fit_transform(cc_apps[col])
        
```

```python
from sklearn.model_selection import train_test_split
cc_apps = cc_apps.drop([11, 13], axis=1)
cc_apps = cc_apps.values

X,y = cc_apps[:,0:12] , cc_apps[:,13]

X_train, X_test, y_train, y_test = train_test_split(X,
                                y,
                                test_size= 0.33,
                                random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
rescaledX_train = scaler.fit_transform(X_train)
rescaledX_test = scaler.fit_transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression ()
logreg.fit(rescaledX_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix
y_pred = logreg.predict(rescaledX_test)
logreg.score(rescaledX_test,y_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
#we can see here that the number of true positives are 10 points higher and the number of true negatives has also increased
#We can also see that the number of false values has only increased by 3. Hence, the values that we dropped could actually be predicted 
#well be using the mean imputation and max char

from sklearn.model_selection import GridSearchCV 
tol=[0.01, 0.001,0.0001]
max_iter=[100, 150,200]
```

```python
param_grid = dict({'tol':tol, 'max_iter':max_iter})
grid_model = GridSearchCV(estimator=logreg, param_grid=param_grid, cv=5)
rescaledX = scaler.fit_transform(X)

# Fit data to grid_model
grid_model_result = grid_model.fit(rescaledX,y)
best_score, best_params = grid_model_result.best_score_, grid_model_result.best_params_
print("Best: %f using %s" % (best_score, best_params))
```
